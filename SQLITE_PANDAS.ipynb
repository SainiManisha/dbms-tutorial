```
import sqlite3
```

```
import pandas as pd
```

```
df=pd.read_csv('https://s3.amazonaws.com/tripdata/JC-201709-citibike-tripdata.csv.zip')
```

```
df
```

```
df.head()
```

```
conn=sqlite3.connect('citibike.db')
```

```
df.to_sql('tripdata',conn)
```

```
sql= 'select * from tripdata'
pd.read_sql(sql, conn)
```

```
sql= 'select stoptime from tripdata'
pd.read_sql(sql, conn)
```

```
sql= 'select  stoptime,tripduration from tripdata'
pd.read_sql(sql, conn)
```

```
sql= 'select * from tripdata GROUP BY bikeid'
pd.read_sql(sql, conn)
```

```
sql='select * from tripdata where tripduration> 800'
pd.read_sql(sql, conn)
```

```
sql='select count(*) from tripdata where tripduration >800'
pd.read_sql(sql, conn)
```

```
sql = 'select bikeid from tripdata where tripduration >800'
pd.read_sql(sql, conn)
```

```
sql='update tripdata set bikeid=4 where tripduration >800'
cur = conn.cursor()
cur.execute(sql)
```
