```
import pandas as pd
import sqlite3
```

The database is named as **"students_data.db"** and it contains a table called **"course"** which has marks of students.

| attribute | description |
| --- | --- |
| admission_no | Admission number of student - unique to each student |
| name | Name of the student |
| section | Whichever section he/she belongs to |
| mse1 | Marks obtained in MSE1 |
| mse2 | Marks obtained in MSE2 |
| assignment | Marks obtained in assignment |
| class_performance | Marks obtained as per class performance |

You are given the dataset and you have to answer a few questions using basic SQL queries (CRUD ops).

```
db = sqlite3.connect('students_data.db')
```

1. Show details of internal marks of all students.

```
query = """
SELECT * FROM course
"""

pd.read_sql(query, db)
```

2. Identify the total number of students in the class.

```
query = """
SELECT COUNT(*) FROM course
"""

print("The total number of students in the class:")
pd.read_sql(query, db)
```

3. The number of students per section (T1, T2).

```
query = """
SELECT section, COUNT(*) FROM course GROUP BY section
"""

print("The number of students per section:")
pd.read_sql(query, db)
```

4. The average marks obtained in `mse1`, `mse2`, `assignment` for all students.

```
query = """
SELECT AVG(mse1), AVG(mse2), AVG(assignment) FROM course
"""

print("The average marks per student:")
pd.read_sql(query, db)
```

5. The total internal marks obtained by each student (internal marks is sum of mse1, mse2, assignment, class_performance)

```
query = """
SELECT *, (mse1 + mse2 + assignment + class_performance) AS total FROM course
"""

print("Total internal marks:")
pd.read_sql(query, db)
```

6. The highest marks in assignment obtained by T1 students and by T2 students.

```
query = """
SELECT section, MAX(assignment) FROM course GROUP BY section
"""

print("Assignment highest:")
pd.read_sql(query, db)
```

7. Students who have failed the course. (total marks obtained is less than 20 marks)

```
query = """
SELECT *, (mse1 + mse2 + assignment + class_performance) AS total FROM course
WHERE total < 20
"""

print("Failed students:")
pd.read_sql(query, db)
```

8. Details of students whose name starts with 'S'.

```
query = """
SELECT * FROM course
WHERE name LIKE 'R%'
"""

print("Students with name starting in 'R':")
pd.read_sql(query, db)
```

9. Move all T2 students to T1 section.

```
query = """
UPDATE course SET section = 'T1' WHERE section = 'T2'
"""

db.execute(query)
db.commit()

pd.read_sql("SELECT * FROM course", db)
```

10. Delete the records of students where `assignment` marks is less than 10.

```
query = """
DELETE FROM course WHERE assignment < 10
"""

db.execute(query)
db.commit()

print("Should be blank table")
pd.read_sql("SELECT * FROM course WHERE assignment < 10", db)
```

11. The total number of student records in the table after all operations.

```
query = """
SELECT COUNT(*) FROM course
"""

print("Number of students left")
pd.read_sql(query, db)
```
